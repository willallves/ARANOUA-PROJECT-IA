2 - Usando as técnicas vistas em laboratório, destaque a fissura na imagem do raio x (Fissura_1.jpg). (2,0 pontos)

```python
import cv2
import numpy as np
from pylab import *
import matplotlib.pyplot as plt
```

```python
path = '/app/introducao_aprendizado_maquina/T2/'
```

```python
def plot_imagens_2(img1, tit1, img2, tit2, res):
    figure(figsize = res)
    plt.subplot(1,2,1), plt.imshow(img1), plt.title(tit1)
    plt.xticks([]),plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(img2), plt.title(tit2)
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
# Carregar a imagem
img = cv2.imread(path + 'Fissura_1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Aplicar um filtro de suavização para reduzir o ruído
img_gaussina_blur = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# Aplicar o operador de Sobel para detecção de bordas
sobel_x = cv2.Sobel(img_gaussina_blur, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(img_gaussina_blur, cv2.CV_64F, 0, 1, ksize=5)

# Calcular a magnitude do gradiente
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Definir um limiar para binarizar a imagem
limiar = 100
img_bin = np.uint8(magnitude > limiar)
```

```python
# Exibir a imagem original e a imagem destacando a fissura
plot_imagens_2(img, 'Original', img_bin, 'Fissura Destacada', (20,20))
```
