1 - Execute os procedimentos para retirada de ru√≠do das 3 imagens anexas - imagem_1, imagem_2 e imagem_3. (3 pontos).

```python
import cv2
from pylab import *
import matplotlib.pyplot as plt
```

```python
path = '/app/introducao_aprendizado_maquina/T2/'
```

```python
def plot_imagens_3(img1, tit1, img2, tit2, img3, tit3, res):
    figure(figsize = res)
    plt.subplot(1,3,1), plt.imshow(img1), plt.title(tit1)
    plt.xticks([]),plt.yticks([])
    plt.subplot(1,3,2), plt.imshow(img2), plt.title(tit2)
    plt.xticks([]),plt.yticks([])
    plt.subplot(1,3,3), plt.imshow(img3), plt.title(tit3)
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
img_1 = cv2.imread(path + 'imagem_1.jpg')
img_2 = cv2.imread(path + 'imagem_2.jpg')
img_3 = cv2.imread(path + 'imagem_3.jpg')
```

```python
plot_imagens_3(img_1, 'original', img_2, 'original', img_3, 'original', (15,15))
```

```python
img_media_blur_1 = cv2.medianBlur(img_1, 7)
img_media_blur_2 = cv2.medianBlur(img_2, 7)
img_media_blur_3 = cv2.medianBlur(img_3, 7)
```

```python
plot_imagens_3(img_media_blur_1, 'filtro_media', img_media_blur_2, 'filtro_media', img_media_blur_3, 'filtro_media', (15,15))
```

```python
img_ruido_1 = img_1 - img_media_blur_1
img_ruido_2 = img_2 - img_media_blur_2
img_ruido_3 = img_3 - img_media_blur_3
```

```python
plot_imagens_3(img_1, 'original', img_media_blur_1, 'filtro_media', img_ruido_1, 'ruido', (15,15))
plot_imagens_3(img_2, 'original', img_media_blur_2, 'filtro_media', img_ruido_2, 'ruido', (15,15))
plot_imagens_3(img_3, 'original', img_media_blur_3, 'filtro_media', img_ruido_3, 'ruido', (15,15))
```
